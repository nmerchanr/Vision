## SLIC: Simple Linear Iterative Clustering (metodo de super pixel).

```python
import cv2
from skimage.segmentation import slic
from skimage.data import astronaut
import matplotlib.pyplot as plt

img = cv2.imread("Panel_0.jpg")

#SLIC
segments1 = slic(img, n_segments=500, compactness=10)
segments2 = slic(img, n_segments=300, compactness=10)
segments3 = slic(img, n_segments=100, compactness=10)



#Visualizacion
plt.figure(1,figsize=(10,5))
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(segments1,cmap = 'gray')
plt.title('segments1'), plt.xticks([]), plt.yticks([])

plt.figure(2,figsize=(10,5))
plt.subplot(121),plt.imshow(segments2,cmap = 'gray')
plt.title('segments2'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(segments3,cmap = 'gray')
plt.title('segments3'), plt.xticks([]), plt.yticks([])


plt.show()
```

## SLIC Superpixel segmentation in scikit image (metodo de super pixel)

```python
from skimage.segmentation import slic, mark_boundaries
#from skimage.data import lena
from skimage.segmentation import slic as slic_wrap 
import matplotlib.pyplot as plt

from skimage.data import astronaut

img = cv2.imread("Panel_3.jpg")

#skimage version\n enforce_connectivity=False', size=9
slic2 = slic(img, 100, 10, sigma=1, enforce_connectivity=False) 

#skimage version\n  enforce_connectivity=True', size=9
slic3 = slic(img, 100, 10, sigma=1, enforce_connectivity=True)



#Visualizacion
plt.figure(1,figsize=(10,5))

plt.subplot(122),plt.imshow(mark_boundaries(img, slic2),cmap = 'gray')
plt.title('skimage version\n enforce_connectivity=False'), plt.xticks([]), plt.yticks([])

plt.show()
```
